**Function to load X dataset.. train as well as test**::

```python
import os
```

```python
Train_directory = "../input/iiitb-ai511-ams-solar-energy-prediction-contest/gefs_train_updated/train/"
Test_directory = "../input/iiitb-ai511-ams-solar-energy-prediction-contest/gefs_test_updated/test"
```

```python
def get_lat_long_range(station):
    min_lat=int(np.floor(data_station[data_station.stid==station].nlat))-31
    max_lat=int(np.ceil(data_station[data_station.stid==station].nlat))-31

    min_long=int(np.floor(data_station[data_station.stid==station].elon+360))-254
    max_long=int(np.ceil(data_station[data_station.stid==station].elon+360))-254
    return (min_lat,max_lat,min_long,max_long)
```

```python
def load_netcdf4(directory,file_name,substr,station):
    print('Collecting', file_name)
    file_path = os.path.join(directory,file_name+substr)
    X = list(Dataset(file_path).variables.values())[-1][:4018]
    X = X.reshape(X.shape[0], 55, 9, 16)
    X = np.mean(X,axis=1)
    min_lat,max_lat,min_long,max_long=get_lat_long_range(station)
    X = X[:,min_lat:max_lat+1,min_long:max_long+1]
    X = X.reshape(X.shape[0],np.prod(X.shape[1:]))

    return X 
```

```python
def stack_netcdf4(directory,features,substr,location):
    for index,feature in enumerate(features):
        if index == 0:
            X = load_netcdf4(directory,feature,substr,location)
        else :
            temp = load_netcdf4(directory, feature ,substr,location)
            X = np.hstack((X,temp))
        
    return X       
            
   
```

```python
def getData_X():
    for index,location in enumerate(location_req):
        print('For location ::', location )
        if (index==0):
            X_train=stack_netcdf4(Train_directory,features,train_substring,location)
        else:
            X_train=np.vstack([X_train,stack_netcdf4(Train_directory,features,train_substring,location)])

    return X_train
```
